**1.3.1.	Data Types:**
Python is a dynamically typed programming language; therefore, data type is assigned to variable on run time based on its value. In python data types are classes and defined variables are their objects.

Some basic data types that are supported in python are listed below;

a.	 **int –** numeric data type represents an integer.

b.	 **float -** numeric data types represent a decimal number.

c.	 **Complex –** numeric data type represents a complex number.

d.	 **Str –** string data type.

e.	 **Bool –** Boolean data type.

Example: 

In an example below, data type is assigned to variable on run time based on its value.


```python
x =2;
print(f"type of variable x is {type(x)}")
```

```python
x =2.2;
print(f"type of variable x is {type(x)}")
```

```python
x ="HPC";
print(f"type of variable x is {type(x)}")
```

Even in some cases, when taking the user input, we can force the variable to be of specific data type. An example is reported below;

```python
user_input = int(input("enter your id:"))
print(f"user id is: {user_input}")
```

```python
user_input = int(input("enter your id:"))
print(f"user id is: {user_input}")
```

**1.3.3.	Flow Control:**

In python, flow control is done using conditional statement and loops, details are reported below;

a.	**if -else statement-** decision making

b.	**if – elif -else statement-** decision making

c.	for loop 

d.	while loop
e.	**break –** exit loop 

f.	**pass --** do nothing

g.	**continue –** skip the execution in current iteration and transfer control to next 
      iteration.
      
Syntax is reporte below;

**if – else statement:**

**if condition1:**  
         #run if condition1 is true  
**else:**  
        #run if condition1 is false 


```python
user_input = int(input("enter code: "))
if user_input == 223:
    print("code is correct")
else:
    print("code is incorrect")
```

**if – elif -else**

**if condition1:**  
	# Run if condition1 is true  
**elif condition2:**  
	# run is condition 2 is true  
**else:**  
    # Run this if none of conditions are true
    
An important point to remember in “if -elif – else”, for multiple elif, if first “elif” condition is true, it will not check the other “elif” in entire chain.

An example is reported below;


```python
user_input = int(input("enter code: "))
if user_input == 223:
    print("code is correct")
elif user_input == 112:
    print("code is correct")
elif user_input == 112:
    print("duplicate code")
else:
    print("code is incorrect")
```

**For loop:**

  Syntax is reported below;
  
for variable in sequence:  
	# run this block


```python
for i in range(10):
    print(i)
```

Use of **“break”** is reported in an example below;

```python
for i in range(10):
    print(i)
    if i == 5:
        break
```

Use of **“continue”** is reported in an example below;

```python
for i in range(10):
    if i == 5:
        continue
    print(i)
    
```

Use of **“pass”** is reported in an example below;

```python
for i in range(10):
    if i == 5:
        pass
    print(i)
```

**Comparison between python & Fortran.**

| Category | Feature | Python | Fortran |
|----------|----------|---------|----------|
| **Data Type** | Typing | Dynamically typed | Static typing |
| | Int | Auto-sized | Fixed kind |
| | Float | Double precision | Real, Double Precision |
| | Complex | Built-in complex | Built-in complex |
| | Boolean | Bool | Logical |
| | String | Flexible `str` | `character(len=...)` |
| **Data Structure** | Array | List, NumPy array | Native array |
| | List | Built-in dynamic list | Not native (use array) |
| | Tuple, Dictionary | Available | Not available |
| **Flow Control** | Conditional | `if`, `elif`, `else` | `if ... then ... else` |
| | Loop (for) | `for i in range()` | `do i = 1,n` |
| | While loop | `while` | `do while` |
| | Break / Continue | Supported | `exit`, `cycle` |


Fortran provides faster execution and better memory control. Fortran is extremely fast for scientific computation and compiler optimization are strong, on the other hand python provides ease of use, huge eco system of libraries for AI.
