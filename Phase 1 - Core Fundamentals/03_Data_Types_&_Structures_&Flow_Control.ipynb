# Data Types, Data Structures, and Flow Control in Python

## Introduction
A scientific program must represent numerical values, strings, logical conditions, collections of data, and repeated operations. Python provides built-in data types, data structures, and flow-control statements that make this kind of organization compact and readable.

In Python, these tools are often flexible and dynamically typed, while Fortran expresses many of the same ideas through explicit declarations, arrays, and structured control constructs.

## 1. Data Types

Python is a dynamically typed language. This means that the type of a value is determined at runtime from the object assigned to a name. More precisely, Python values are objects, and variables are names that refer to those objects.

Some basic Python data types are:

* `int`: integer numbers;
* `float`: floating-point numbers;
* `complex`: complex numbers;
* `str`: text strings;
* `bool`: Boolean values, either `True` or `False`.

````python
x = 2
print(f"x = {x}, type: {type(x)}")
````

````python
x = 2.2
print(f"x = {x}, type: {type(x)}")
````

````python
x = "HPC"
print(f"x = {x}, type: {type(x)}")
````

````python
x = 1 + 2j
print(f"x = {x}, type: {type(x)}")
````

````python
x = True
print(f"x = {x}, type: {type(x)}")
````

The same name `x` can be associated with objects of different types during execution. This is convenient during prototyping, but it also means that type-related mistakes may appear only when the program is executed.

### Type Conversion

In some cases, a value must be converted explicitly from one type to another. This often happens when values are read from files or from user input, because they may initially be represented as strings.

````python
raw_id = "123"
user_id = int(raw_id)

print(f"raw_id: {raw_id}, type: {type(raw_id)}")
print(f"user_id: {user_id}, type: {type(user_id)}")
````

## 2. Data Structures

A data structure is a way of organizing and storing data. Python provides several built-in data structures that are useful for small datasets, configurations, and workflow organization.

The most common ones are:

* `list`: ordered and mutable collection;
* `tuple`: ordered and immutable collection;
* `set`: unordered collection of unique elements;
* `dict`: mutable collection of key-value pairs.

### Lists

A list is an ordered and mutable collection. Its elements can be accessed by index, and new elements can be added after the list has been created.

````python
course = ["physics", "maths", "english"]

print(f"First course: {course[0]}")
print(f"Second course: {course[1]}")
print(f"Third course: {course[2]}")

course.append("italiano")
print(f"Updated list: {course}")
````

### Tuples

A tuple is ordered but immutable. This means that its elements can be accessed by index, but the tuple cannot be modified by adding or removing elements.

````python
course = ("physics", "maths", "english")

print(f"First course: {course[0]}")
print(f"Second course: {course[1]}")
print(f"Third course: {course[2]}")

course.append("italiano")
````

### Sets

A set is an unordered collection of unique elements. Sets are useful when duplicated values must be removed or when membership tests are important.

````python
course = {"physics", "maths", "english", "physics"}

print(f"Courses: {course}")

course.add("italiano")
print(f"Updated courses: {course}")
````

### Dictionaries

A dictionary stores key-value pairs. This is useful when each value has a meaningful name or identifier.

````python
temperature_sensor = {
    "ID": "001",
    "value": 23,
    "unit": "celsius",
}

print(temperature_sensor)
print(f"Sensor value: {temperature_sensor['value']} {temperature_sensor['unit']}")
````

````python
temperature_sensors = {
    1: {"ID": "001", "value": 23, "unit": "celsius"},
    2: {"ID": "002", "value": 22, "unit": "celsius"},
}

print(temperature_sensors[1])
````

## 3. Flow Control

Flow control determines which instructions are executed and how many times they are repeated. In Python, the main flow-control tools are conditional statements and loops.

The most common constructs are:

* `if`, `elif`, `else` for decision making;
* `for` loops for iterating over a sequence;
* `while` loops for repeating a block while a condition is true;
* `break` to exit a loop;
* `continue` to skip the current iteration;
* `pass` to do nothing when a statement is syntactically required.

### `if`, `elif`, and `else`

Conditional statements allow the program to execute different instructions depending on logical conditions.

````python
code = 223

if code == 223:
    print("Code is correct")
else:
    print("Code is incorrect")
````

````python
code = 112

if code == 223:
    print("Code 223 is correct")
elif code == 112:
    print("Code 112 is correct")
elif code == 112:
    print("This branch is never reached")
else:
    print("Code is incorrect")
````

When using an `if`-`elif`-`else` chain, conditions are checked in order. As soon as one condition is true, the corresponding block is executed and the remaining branches are skipped.

### `for` Loops

A `for` loop repeats a block of instructions for each element of a sequence. The function `range()` is often used to generate integer sequences.

````python
for i in range(10):
    print(i)
````

The instruction `break` exits the loop immediately.

````python
for i in range(10):
    print(i)
    if i == 5:
        break
````

The instruction `continue` skips the rest of the current iteration and moves to the next one.

````python
for i in range(10):
    if i == 5:
        continue
    print(i)
````

The instruction `pass` does nothing. It can be used as a placeholder when Python syntax requires a statement.

````python
for i in range(5):
    if i == 2:
        pass
    print(i)
````

### `while` Loops

A `while` loop repeats a block of instructions while a condition remains true.

````python
x = 0

while x < 5:
    print(f"value of x: {x}")
    x += 1
````

## 4. Comparison with Fortran

Fortran uses explicit declarations for variables. With `IMPLICIT NONE`, variables must be declared before they are used. The same basic information represented dynamically in Python is therefore written explicitly in Fortran.

```fortran
integer :: n
real :: x
complex :: z
logical :: flag
character(len=20) :: label
```

Fortran also has structured control constructs. Conditional execution is written with `IF`, `ELSE IF`, `ELSE`, and `END IF`:

```fortran
if (code == 223) then
    print *, "Code is correct"
else
    print *, "Code is incorrect"
end if
```

A loop over an integer counter is written with `DO`:

```fortran
do i = 1, 10
    print *, i
end do
```

The Fortran instructions corresponding to Python `break` and `continue` are `EXIT` and `CYCLE`, respectively.

Python lists and Fortran arrays should not be confused. A Python list is a flexible container that can store references to different kinds of objects. A Fortran array is a homogeneous collection of variables of the same type, stored contiguously in memory and accessed with numerical indices. In scientific Python, NumPy arrays provide the array structure used for efficient numerical work.

## 5. Pros and Cons from an HPC Perspective

| Aspect | Python | Fortran |
| --- | --- | --- |
| Typing | Dynamic, checked at runtime | Explicit declarations checked before execution |
| Basic data | `int`, `float`, `complex`, `bool`, `str` | `INTEGER`, `REAL`, `COMPLEX`, `LOGICAL`, `CHARACTER` |
| Data structures | Built-in lists, tuples, sets, dictionaries | Native arrays, derived types, modules |
| Loops | `for`, `while`, `break`, `continue` | `DO`, `DO WHILE`, `EXIT`, `CYCLE` |
| Numerical arrays | Built-in lists are flexible but not optimized numerical arrays | Native homogeneous arrays with explicit type and shape |

**Strengths of Python in this context:**

* compact syntax;
* flexible data structures;
* readable parameter organization through dictionaries;
* convenient for prototyping and small workflows.

**Limitations:**

* dynamic typing can hide errors until runtime;
* Python lists are not equivalent to Fortran arrays;
* Python loops can become inefficient for large numerical datasets;
* efficient scientific computation usually requires specialized libraries such as NumPy.

## 6. Summary

Python provides simple and expressive tools for representing values, grouping data, and controlling program execution. Data types describe individual values, data structures organize collections of values, and flow-control statements determine the order in which instructions are executed.

Compared with Fortran, Python is more flexible and compact, while Fortran is more explicit about types, arrays, and control structures. In an HPC context, Python built-in structures are useful for organization and prototyping, but large numerical computations require more efficient array-based tools, such as NumPy.
