# Functions and Classes in Python: OOP for Scientific and HPC Code

## Introduction
A well written code must be readable, reusable, testable, and easy to extend. This is especially important in High-Performance Computing (HPC), where programs often combine numerical computations, input/output, parameter management, parallel execution, and post-processing.

Python provides two central tools for organizing code:

1. **Functions**, which group instructions into reusable computational blocks.
2. **Classes**, which combine data and operations into objects, following the Object-Oriented Programming (OOP) paradigm.

The key point from an HPC perspective is that Python functions and classes are excellent for structuring the simulation, but the most expensive numerical operations should usually be performed with optimized libraries such as NumPy, SciPy, or compiled code. In parallel programs, mpi4py can then be used to distribute the work and exchange data between processes.

## 1. Functions: Reusable Computational Blocks

A function is defined using the `def` keyword, followed by the function name, a list of parameters, and an indented body. The first statement of a function body can be a **docstring**, which documents the purpose of the function.

Functions are fundamental in scientific code because they allow us to separate the mathematical model from the rest of the program. For example, the kinetic energy formula

$$E_k = \frac{1}{2}mv^2$$

can be written as a reusable Python function.

````python
def kinetic_energy(mass, velocity):
    """Return the kinetic energy of a particle."""
    return 0.5 * mass * velocity**2


m = 2.0
v = 3.0

print(f"Kinetic energy: {kinetic_energy(m, v)}")
````

### Python syntax used

* `def kinetic_energy(...)`: defines a new function.
* `mass, velocity`: function parameters.
* `return`: sends a value back to the caller.
* `"""..."""`: docstring, useful for documenting scientific routines.

Python functions are dynamically typed: the function does not require us to declare in advance whether `mass` and `velocity` are integers, floats, arrays, or other compatible objects. This makes the code flexible, but also means that type checks and operation dispatch happen at runtime.

## 2. Comparison with Fortran Procedures

In Fortran, code modularity is built around **procedures**. A procedure is a callable block of code that performs a specific task. In particular, Fortran distinguishes between:

* **functions**, which return a single value and can be used inside expressions;
* **subroutines**, which are invoked with a `CALL` statement and can return multiple results through their arguments;
* **modules**, which store data, definitions, derived types, procedures, and interfaces for reuse in other program units.

The Fortran version of the kinetic energy function has a more explicit structure: the return type and argument types are declared, and the `intent(in)` attribute specifies that the arguments are input-only.

```fortran
real function kinetic_energy(mass, velocity)
    implicit none
    real, intent(in) :: mass, velocity
    kinetic_energy = 0.5 * mass * velocity**2
end function kinetic_energy
```

This explicitness gives the compiler more information about the code before execution. Python, instead, favors compact syntax and runtime flexibility.

## 3. Functions in HPC: Clarity vs. Overhead

Functions improve readability and make code easier to test. However, in pure Python, calling a function inside a very large numerical loop can introduce significant overhead. Each call requires the interpreter to manage arguments, stack frames, dynamic type dispatch, and return values.

This does not mean that functions should be avoided. It means that in HPC Python code, functions are best used to organize the workflow and to wrap vectorized or compiled operations.

````python
import time
import numpy as np


def scalar_kinetic_energy(mass, velocity):
    return 0.5 * mass * velocity**2


N = 1_000_000
masses = np.ones(N, dtype=np.float64) * 2.0
velocities = np.linspace(0.0, 10.0, N)

# Pure Python loop: many calls to a Python function
start_time = time.time()
energies_loop = [scalar_kinetic_energy(masses[i], velocities[i]) for i in range(N)]
loop_time = time.time() - start_time

# Vectorized NumPy expression: one array operation executed in compiled code
start_time = time.time()
energies_vectorized = 0.5 * masses * velocities**2
vectorized_time = time.time() - start_time

print(f"Python loop time:        {loop_time:.4f} s")
print(f"NumPy vectorized time:   {vectorized_time:.4f} s")
print(f"Speedup:                {loop_time / vectorized_time:.1f}x")
````

In the previous example, the first implementation computes the kinetic energy one element at a time and calls a Python function at each iteration. The second implementation applies the same formula directly to the full NumPy arrays. This keeps the mathematical expression readable while moving the repeated element-wise work into optimized array operations.

## 4. Classes and Object-Oriented Programming

A **class** defines a new type of object. In Python, classes group together:

* **attributes**, which store data;
* **methods**, which are functions associated with the object.

Python classes provide the standard features of Object-Oriented Programming with a relatively compact syntax. They make it possible to define new object types, store data inside each object, and associate operations with that data.

A simple example is a `Particle` class. The special method `__init__` initializes the object when it is created. The first parameter of instance methods is conventionally called `self`, and it refers to the current object.

````python
class Particle:
    def __init__(self, mass, velocity):
        self.mass = mass
        self.velocity = velocity

    def kinetic_energy(self):
        return 0.5 * self.mass * self.velocity**2


electron = Particle(mass=9.109e-31, velocity=2.0e6)

print(f"Mass: {electron.mass:.3e} kg")
print(f"Velocity: {electron.velocity:.3e} m/s")
print(f"Kinetic energy: {electron.kinetic_energy():.3e} J")
````

### Python syntax used

* `class Particle`: defines a new class.
* `__init__`: constructor-like initialization method.
* `self.mass`, `self.velocity`: object attributes.
* `kinetic_energy(self)`: method operating on the object state.

This style is very natural for modeling physical entities. A particle has data, such as mass and velocity, and operations, such as the computation of kinetic energy.

## 5. Python Classes and Fortran Derived Types

The closest Fortran concept is the **derived type**. Derived types allow the programmer to create structured data by grouping heterogeneous components into a single object.

For example, derived types define structured objects with components accessed through the `%` operator:

```fortran
type :: particle
    real :: mass
    real :: x
end type particle

type(particle) :: p
p%mass = 1.0
p%x = 0.5
```

A Python object uses dot notation instead:

```python
p = Particle(mass=1.0, velocity=0.5)
p.mass
```

The similarity is that both languages allow us to group related data. The main difference is that in Python the same `class` block usually contains both the data attributes and the methods that operate on them, instead in Fortran a derived type mainly defines the data structure, while the operations on that data are usually written as separate functions or subroutines, often organized inside modules.

## 6. Encapsulation and Validation

Classes are also useful because they let us protect the consistency of the object state. In scientific codes, invalid parameters can silently produce meaningless results. Python exceptions provide a direct way to stop the program when the input is physically or numerically invalid.

Exceptions are used to handle errors detected during execution. We can use `raise` to signal that a value is not acceptable.

````python
class SimulationConfig:
    def __init__(self, n_particles, dt):
        if n_particles <= 0:
            raise ValueError("n_particles must be positive")
        if dt <= 0.0:
            raise ValueError("dt must be positive")

        self.n_particles = n_particles
        self.dt = dt


config = SimulationConfig(n_particles=100_000, dt=0.01)

print(f"Number of particles: {config.n_particles}")
print(f"Time step: {config.dt}")
````

## 7. HPC Design Pattern: Objects Outside, Arrays Inside

A common mistake is to represent a large physical system as a Python list of many small objects. This is readable, but inefficient: each object carries Python-level overhead, and the data are not stored as one compact numerical block.

For HPC, a better pattern is:

* use classes for configuration, validation, I/O, and high-level organization;
* use NumPy arrays for large numerical data;
* use vectorized operations, SciPy routines, or compiled libraries for heavy computation;
* use mpi4py to distribute array-based work across MPI processes.

This combines the readability of object-oriented Python with the performance of contiguous numerical arrays.

````python
class ParticleSystem:
    def __init__(self, n_particles):
        if n_particles <= 0:
            raise ValueError("n_particles must be positive")

        self.n_particles = n_particles
        self.positions = np.zeros((n_particles, 3), dtype=np.float64)
        self.velocities = np.random.default_rng(seed=42).normal(size=(n_particles, 3))

    def speed(self):
        return np.linalg.norm(self.velocities, axis=1)

    def advance(self, dt):
        self.positions += self.velocities * dt


system = ParticleSystem(n_particles=100_000)
system.advance(dt=0.01)

print(f"Mean speed: {system.speed().mean():.4f}")
print(f"Positions array shape: {system.positions.shape}")
````

The class organizes the simulation state, but the particle coordinates and velocities are stored in NumPy arrays. The method `advance` does not loop over particles in Python: it updates the whole array using a vectorized operation.

This design is much closer to an HPC style than creating 100,000 individual `Particle` objects.

## 8. Pros and Cons from an HPC Perspective

| Aspect | Python functions/classes | Fortran procedures/derived types |
| --- | --- | --- |
| Syntax | Compact and expressive | More explicit and verbose |
| Typing | Dynamic, checked at runtime | Explicit declarations checked by the compiler |
| Modularity | Functions, classes, modules | Functions, subroutines, modules |
| Structured data | Classes with attributes and methods | Derived types with components, often combined with modules/procedures |
| Performance | Function calls and objects add runtime overhead | Compiler has more static information for optimization |
| HPC use | Excellent for orchestration, prototyping, analysis | Excellent for computationally intensive numerical code |

**Strengths of Python OOP in HPC:**

* improves readability and organization of scientific codes;
* reduces development time;
* makes configuration and validation easier;
* integrates naturally with NumPy, SciPy, mpi4py, plotting tools, and notebooks.

**Limitations:**

* Python objects have memory overhead;
* method calls are slower than compiled procedure calls;
* dynamic typing and runtime dispatch reduce raw performance;
* object-per-element designs are inefficient for large numerical datasets.

## 9. Summary

Functions and classes are essential tools for writing clear scientific Python code. Functions encapsulate reusable calculations, while classes allow us to represent objects such as particles, grids, simulation parameters, or numerical solvers.

From the HPC point of view, the most important rule is not to confuse code organization with numerical performance. Python OOP is useful for organizing the high-level structure of the application. However, when the code operates on large numerical datasets, it is usually more efficient to store the data in arrays and use vectorized operations or optimized libraries.
