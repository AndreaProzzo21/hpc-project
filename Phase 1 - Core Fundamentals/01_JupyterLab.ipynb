**1.1.	 Jupyter Lab:**
Jupyter Lab is a web based interactive environment, used for coding, documentation and data analysis within the single workspace.

Main components of jupyter lab are;

**1.1.1	Browser:** It is the front end for user interface. It connects to jupyter server via HTTP / WebSocket.

**1.1.2	Jupyter Server:** The server runs on either a local machine or in remote mode. It is connected to browser via HTTP. On the other hand, jupyter server is connected to kernel via ZeroMQ protocol.

**1.1.3	Kernel:** A kernel is an OS based process. It is the main execution engine and runs the code and return the results to jupyter server. 

**1.1.4	Notebook:**  It contains cells, code is written in different cell, variable declare in one cell can be access in the other cell within the same notebook. For different notebooks, each notebook gets its own dedicated kernel, thus each notebook runs in a fully isolated kernel. In notebook user can run selected cells, or run all cells, when run all cells, jupyter lab runs the cell from top to bottom.

**1.1.5	Filesystem:**  File system can be local or remote, depending upon how jupyter server is launched. When the command open (‘file.txt’,’w’) is executed, kernel ask the local operating system to create the file on server’s hard drive. 

**Some limitations of Jupyter Lab in HPC:** 

a.	In a notebook, if a cell is long enough it blocks other cells, thus no cell level parallelism.

b.	Large data set remain in memory, can crash kernels, only way to restart the kernel.

c.	Browser dependency.

d.	Long running production jobs: kernel may get disconnected, browser tabs closure risk.

When using jupyter lab, MPI4 can be used for parallel computing using python.  Slrum scheduler can be used to submit the job.
