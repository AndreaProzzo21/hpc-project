# JupyterLab as an Interactive Environment for Scientific Computing

## Introduction
JupyterLab is a web-based interactive environment used for coding, documentation, and data analysis within a single workspace. It is particularly useful in scientific projects because code, explanations, plots, and results can be developed together in the same notebook.

## 1. Main Components of JupyterLab

JupyterLab is based on different components that work together during code execution.

**Browser.** The browser provides the graphical user interface. It connects to the Jupyter server through HTTP and WebSocket connections.

**Jupyter Server.** The server can run on a local machine or on a remote system. It receives requests from the browser and communicates with the kernel.

**Kernel.** The kernel is an operating-system process that executes the code. It receives the code from the Jupyter server, runs it, and sends the results back.

**Notebook.** A notebook contains cells. Code written in one cell can define variables that are then available in later cells of the same notebook, as long as they are executed by the same kernel. Different notebooks can use different kernels, keeping their executions separated.

**Filesystem.** The filesystem can be local or remote, depending on where the Jupyter server is running. For example, when a command such as `open("file.txt", "w")` is executed, the kernel asks the operating system to create the file in the filesystem available to the server.

## 2. Code Execution in a Notebook

A notebook is organized into cells. The user can execute a single selected cell or run all cells in order. When all cells are executed, JupyterLab runs them from top to bottom.

This execution model is convenient for interactive work, but it also requires attention. Since variables remain stored in the kernel memory, the result of a cell may depend on cells that were executed earlier. For this reason, a notebook should be periodically restarted and executed from the beginning to check that the workflow is reproducible.

## 3. Limitations in an HPC Context

JupyterLab is useful for prototyping, documentation, and analysis, but it also has limitations when used in an HPC context:

* in a single notebook connected to one kernel, a long-running cell occupies that kernel, so other cells in the same notebook cannot execute independently until it finishes;
* large datasets can remain in memory and may cause the kernel to crash;
* long-running jobs should not depend on an open browser tab, because the connection may be interrupted.

## 4. JupyterLab and Parallel Computing

JupyterLab can still be useful in parallel-computing workflows. For example, `mpi4py` can be used to write MPI-based Python programs, while a scheduler such as Slurm can be used to submit jobs on an HPC system.

In this kind of workflow, JupyterLab is often more appropriate for preparing code, testing small examples, documenting the analysis, and visualizing results. Longer production jobs are usually executed through scripts and submitted to the scheduler.

## 5. Use of JupyterLab in This Project

In this project, JupyterLab will mainly be used as an interactive editor for combining short theoretical explanations with small Python examples. This is useful for introducing the main topics of the project, such as basic data types and data structures, flow control, file I/O, functions and classes, NumPy, SciPy, and the basic use of `mpi4py`.

For heavier computations, especially parallel runs and the final case study, the code will be executed in the standard way from the terminal or on Lyra. As in the course exercises, longer runs should be submitted through the scheduler rather than executed directly inside a notebook.

## 6. Summary

JupyterLab provides an interactive environment where code, text, and results can be combined in a single workspace. Its browser-server-kernel architecture makes it flexible for local or remote use. In HPC, it is very useful for exploration and analysis, while long computations are better managed through scripts and job schedulers.
